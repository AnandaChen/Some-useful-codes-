```python
n,p1,p2=100,0.5,0.5
#vertex=[[0 for i in range(0,4)] for j in range(0,n)]
#vertex[][0] :value
#vertex[][1]:tail
#vertex[][2]:head

def make(vertex,rightmost_v,k,value):
    #aa=k
    if value>=vertex[rightmost_v][0]:
        vertex[k][0]=round(value,2)
        vertex[k][1]=rightmost_v
        #rightmost_v=k
        return(k)
    rr=rightmost_v
    while vertex[rr][0]>value:
        aa=rr
        rr=vertex[aa][1]
    vertex[k][0]=round(value,2)
    vertex[k][1]=rr
    vertex[aa][1]=k
    return(rightmost_v)
    
vertex=[[0 for i in range(0,3)] for j in range(0,n+10)]


def find(vertex,nn,rightmost_v):
    count=nn
    gg=rightmost_v
    while count>1:
        gg=vertex[gg][1]
        count-=1
    return(gg)
    


def IBMM_4_4(p1,p2,x,n,vertex):    
    rightmost_v=1
    for i in range(2,n+2):
        xi=1
        xi_2=n+100
        v_1,v_1,vv=0,0,0
        
        pd=True
        random_number=random.rand()
        while random_number>p1:            
            if (random_number>p1)and(random_number<=p1+p2)and pd: 
                xi_2=xi
                pd=False
            xi+=1
            random_number=random.rand()         
        vertex[i][2]=xi #mark_1
        vertex[i][3]=i  #maek_2
        vertex[i][4]=xi_2 #mark_3
        if xi>i-1 and xi_2>i-1:
            rightmost_v=make(vertex,rightmost_v,i,0) 
            continue
        if xi>i-1 and xi_2<=i-1:
            v_2=vertex[find(vertex, xi_2,rightmost_v)][0]+x
            rightmost_v=make(vertex,rightmost_v,i,max(v_2,0)) 
            continue
        if xi<=i-1 and xi_2>i-1:
            v_1=vertex[find(vertex, xi,rightmost_v)][0]+1
            rightmost_v=make(vertex,rightmost_v,i,max(v_1,0)) 
            continue
        v_1=vertex[find(vertex, xi,rightmost_v)][0]+1
        v_2=vertex[find(vertex, xi_2,rightmost_v)][0]+x
        v_k=max(v_1,v_2,0)
        
        rightmost_v=make(vertex,rightmost_v,i,v_k)     
    return(rightmost_v)     
n=100
vertex=[[0 for i in range(0,5)] for j in range(0,n+10)]        
IBMM_4_4(0.5,0.5,-0.8,n,vertex)    
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generate 2 2d grids for the x & y bounds
y, x = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))

z = (1 - x / 2. + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
# x and y are bounds, so z should be the value *inside* those bounds.
# Therefore, remove the last value from the z array.
z = z[:-1, :-1]
z_min, z_max = -np.abs(z).max(), np.abs(z).max()

fig, ax = plt.subplots()

c = ax.pcolormesh(x, y, z, cmap='RdBu', vmin=z_min, vmax=z_max)
ax.set_title('pcolormesh')
# set the limits of the plot to the limits of the data
ax.axis([x.min(), x.max(), y.min(), y.max()])
fig.colorbar(c, ax=ax)

plt.show()
```
